## Geting Started

```julia
using Turing
using StatsPlots
```

我们定义一个具有未知均值和方差的简单正态分布模型。

```julia
@model function gdemo(x, y)
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))
    x ~ Normal(m, sqrt(s²))
    return y ~ Normal(m, sqrt(s²))
end
```

我们使用Hamiltonian Monte Carlo进行采样

```julia
chn = sample(gdemo(1.5, 2), NUTS(), 1000)
```

```julia
plot(chn)
```

在这种情况下，因为我们使用normal-inverse gamma分布作为共轭先验，我们可以得到其均值和方差的解析解：

Gamma分布：
$$
f(x; \alpha, \theta) = \frac{x^{\alpha-1} e^{-x/\theta}}{\Gamma(\alpha) \theta^\alpha},
\quad x > 0
$$
Inverse-Gamma分布：
$$
f(x; \alpha, \theta) = \frac{\theta^\alpha x^{-(\alpha + 1)}}{\Gamma(\alpha)}
e^{-\frac{\theta}{x}}, \quad x > 0
$$
其中$\alpha$为shape参数，$\theta$为scale参数。

```julia
s² = InverseGamma(2, 3)
m = Normal(0, 1)
data = [1.5, 2]
x_bar = mean(data)
N = length(data)

mean_exp = (m.σ * m.μ + N * x_bar) / (m.σ + N)
```

```julia
updated_alpha = shape(s²) + (N / 2)
updated_beta = scale(s²) + (1 / 2) * sum((data[n] - x_bar)^2 for n in 1:N) + (N * m.σ) / (N + m.σ) * ((x_bar)^2) / 2
variance_exp = updated_beta / (updated_alpha - 1)
```

我们可以对得到的分布进行采样

```julia
function sample_posterior(alpha, beta, mean, lambda, iterations)
    samples = []
    for i in 1:iterations
        sample_variance = rand(InverseGamma(alpha, beta), 1)
        sample_x = rand(Normal(mean, sqrt(sample_variance[1]) / lambda), 1)
        samples = append!(samples, sample_x)
    end
    return samples
end
analytical_samples = sample_posterior(updated_alpha, updated_beta, mean_exp, 2, 1000);
```

```julia
density(analytical_samples; label="Posterior (Analytical)")
density!(chn[:m]; label="Posterior (HMC)")
```

## Probabilistic Programming in Thirty Seconds

```julia
using Turing, StatsPlots, Random

# 硬币正面朝上的真实概率
p_true = 0.5

Ns = 0:100

# 从伯努利分布中抽样
Random.seed!(12)
data = rand(Bernoulli(p_true), last(Ns))

# 声明我们的Turing模型
@model function coinflip(y)
    p ~ Beta(1, 1)
    
    N = length(y)
    for n in 1:N
        y[n] ~ Bernoulli(p)
    end
end

# 设置HMC采样
iterations = 1000
ϵ = 0.05
τ = 10
# 开始采样
chain = sample(coinflip(data), HMC(ϵ, τ), iterations)

histogram(chain[:p])
```

## Guide

`x~distr`，其中`x`为一个符号并且`distr`为分布。如果`x`未定义，则在模型内部放入一个叫`x`的随机变量，服从`distr`分布。如果`x`是定义的，那么其为观测值。似然通过`logpdf(distr,y)`获得。

我们的采样方法及其参数有：
+ IS(Importance Sampling)
+ SMC(Sequential Monte Carlo)：粒子的数量
+ PG(Partial Gibbs)：粒子的数量，迭代的次数
+ HMC(Hamiltonian Monte Carlo)：leapfrog step size，leapfrog step numbers
+ Gibbs：component sampler 1, component sampler2, ...
+ HMCDA(Hamiltonian Monte Carlo with Dual Averaging): total leapfrog length, target accept ratio
+ NUTS(No-U-Turn Sampler): number of adaptation(optional), target accept ratio

### Modelling Syntax Explained
基本语法

定义模型
~~~julia
@model function model_name(arg_1, arg_2)
    ...
end
~~~

采样
~~~julia
model_func = model_name(1, 2)
chn = sample(model_func, HMC(..))
~~~

结果
~~~julia
var_1 = mean(chn[:var_1])
~~~

如果我们想要提取变量"x[1]","x[2]","x[3]"的结果，我们可以使用`group(chn, :x)`的方法。

定义模型时语句的顺序非常重要，比如，下列模型可以工作：
~~~julia
@model function model_function(y)
    s ~ Poisson(1)
    y ~ Normal(s, 1)
    return y
end

sample(model_function(10), SMC(), 100)
~~~
但是交换顺序后的下列模型不可以：
~~~julia
@model function model_function(y)
    y ~ Normal(s, 1)
    s ~ Poisson(1)
    return y
end

sample(model_function(10), SMC(), 100)
~~~

### Sampling Multiple Chains

#### Multithreaded sampling

我们可以采用下列方法进行多线程采样，但是得事先开启[Julia的多线程](https://docs.julialang.org/en/v1/manual/multi-threading/#man-multithreading)。

```julia
@model function gdemo(x)
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))

    for i in eachindex(x)
        x[i] ~ Normal(m, sqrt(s²))
    end
end

model = gdemo([1.5, 2.0])

# Sample four chains using multiple threads, each with 1000 samples.
sample(model, NUTS(), MCMCThreads(), 1000, 4)
```

### Sampling from an Unconditional Distribution (The Prior)

Turing允许我们从声明的模型的先验中采样。

```julia
@model function gdemo(x, y)
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))
    x ~ Normal(m, sqrt(s²))
    y ~ Normal(m, sqrt(s²))
    return x, y
end
```

gdemo有两个参数，如果我们将这两个参数设置为missing，那么就会认为没有数据，就会从先验中采样。

```julia
g_prior_sample = gdemo(missing, missing)
g_prior_sample()
```

### Sampling from a Conditional Distribution(The Posterior)
### Treating observations as random variables

输入值为missing的被认为是参数，也称为随机变量，所以将输入变量设置为missing我们可以将其当做参数进行采样。

```julia
@model function gdemo(x, ::Type{T} = Float64) where {T} # 这里应该指的是x的默认类型为Float64
    if x === missing
        x = Vector{T}(undef, 2)
    end
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))
    for i in eachindex(x)
        x[i] ~ Normal(m, sqrt(s²))
    end
end
model = gdemo(missing)
c = sample(model, HMC(0.01, 5), 500)
```

值不为空的则认为是观测值。

```julia
@model function gdemo(x)
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))
    for i in eachindex(x)
        x[i] ~ Normal(m, sqrt(s²))
    end
end
model = gdemo([missing, 2.4])
c = sample(model, HMC(0.01, 5), 500)
```

### Default Values

```julia
@model function generative(x = missing, ::Type{T} = Float64) where {T <: Real}
    if x === missing
        x = Vector{T}(undef, 10)
    end
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))
    for i in 1:length(x)
        x[i] ~ Normal(m, sqrt(s²))
    end
    return s², m
end
m = generative()
chain = sample(m, HMC(0.01, 5), 1000)
```

### Variable Types and Type Parameters

随机变量的向量(或矩阵)应该与其先验分布的类型匹配，`<:Integer`对于离散分布，`<:AbstractFloat`对于连续分布。如果要采用Hamiltonian采样器，向量的类型必须为`Real`或`Real`的子类型或者类型参数`T`定义为`function gdemo(x, ::Type{T} = Float64) where {T}`。类似地，当使用particle采样器时，Julia的变量应该为`Array`或者在函数开始声明`function gdemo(x, ::Type{T} = Vector{Float64}) where {T}`。

### Querying Probabilities from Model or Chain

```julia
@model function gdemo0(x)
    s ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s))
    x ~ Normal(m, sqrt(s))
end
```

### Maximum likelihood and Maximum a posterior estimates

```julia
using Optim

@model function gdemo(x)
    s² ~ InverseGamma(2, 3)
    m ~ Normal(0, sqrt(s²))
    
    for i in eachindex(x)
        x[i] ~ Normal(m, sqrt(s²))
    end
end
```

```julia
data = [1.5, 2.0]
model = gdemo(data)
```

```julia
# MLE
mle_estimate = optimize(model, MLE())

# MAP
map_estimate = optimize(model, MAP())
```

默认采用LBFGS优化算法，当然我们也可以修改优化算法：
~~~julia
# Use NelderMead
mle_estimate = optimize(model, MLE(), NelderMead())

# Use SimulatedAnnealing
mle_estimate = optimize(model, MLE(), SimulatedAnnealing())

# Use ParticleSwarm
mle_estimate = optimize(model, MLE(), ParticleSwarm())

# Use Newton
mle_estimate = optimize(model, MLE(), Newton())

# Use AcceleratedGradientDescent
mle_estimate = optimize(model, MLE(), AcceleratedGradientDescent())
~~~
有些优化算法可能需要更多的迭代次数，可以通过
~~~julia
mle_estimate = optimize(model, MLE(), Newton(), Optim.Options(iterations=10_000, allow_f_increases=true))
~~~

### Beyond the Basics

#### Compositional Sampling Using Gibbs

```julia
@model function simple_choice(xs)
    p ~ Beta(2, 2)
    z ~ Bernoulli(p)
    for i in 1:length(xs)
        if z == 1
            xs[i] ~ Normal(0, 1)
        else
            xs[i] ~ Normal(2, 1)
        end
    end
end

simple_choice_f = simple_choice([1.5, 2.0, 0.3])

chn = sample(simple_choice_f, Gibbs(HMC(0.2, 3, :p), PG(20, :z)), 1000)
```

#### Working with filldist and arraydist

##### Constructing product distribution with filldist

```julia
@model function demo(x, g)
    k = length(unique(g))
    a ~ filldist(Exponential(), k)  # 等同于 = Product(fill(Exponential(), k))，Product产生独立同分布的概率，此语句相当于产生k个独立同分布的Exponential
    mu = a[g]
    x .~ Normal.(mu)
end
```

##### Constructing product distributions with arraydist

```julia
@model function demo(x, g)
    k = length(unique(g))
    a ~ arraydist([Exponential(i) for i in 1:k])
    mu = a[g]
    x .~ Normal.(mu)
end
```

## Advanced Usage

### How to Define a Customized Distribution

```julia
using Random
```

1. 定义分布类型

```julia
struct CustomUniform <: ContinuousUnivariateDistribution end
```

2. 实现采样和log-pdf的计算

```julia
# 在[0,1]之间采样
Distributions.rand(rng::AbstractRNG, d::CustomUniform) = rand(rng)
# p(x) = 1 -> log p(x) = 0
Distributions.logpdf(d::CustomUniform, x::Real) = zero(x)
```

3. 定义辅助函数

有的采样器，例如HMC，需要先验的值域没有界限，所以我们需要将值域从$[0,1]$转换到$\mathbb{R}$

```julia
Bijectors.bijector(d::CustomUniform) = Logit(0., 1.)
```

另外我们也可以定义分布的最大和最小值

```julia
Distributions.minimum(d::CustomUniform) = 0
Distributions.maximum(d::CustomUniform) = 1
```

### Update the accumulated log probability in the model definition

我们可以采用宏(macro)来手动增加累积log概率。

```julia
myloglikelihood(x, μ) = loglikelihood(Normal(μ, 1), x)

@model function demo(x)
    μ ~ Normal()
    Turing.@addlogprob! myloglikelihood(x, μ)
end
```

可以用来拒绝采样

```julia
using LinearAlgebra
```

```julia
@model function demo(x)
    m ~ MvNormal(zero(x), I)
    if dot(m, x) < 0
        Turing.@addlogprob! -Inf
        return
    end
    x ~ MvNormal(m, I)
    return
end
```
